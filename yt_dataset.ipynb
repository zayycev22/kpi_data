```python
import pandas as pd
from easyocr import Reader
from tqdm import tqdm
from catboost import CatBoostRegressor
ALLOWED_FILES = ('.png', ".jpg", ".jpeg")
```

```python
reader = Reader(lang_list=['ru', 'en'], gpu=True)
```

```python
yt1_df = pd.read_excel("train_dataset/yt/yt1.xlsx")
yt1_df
```

```python
yt2_df = pd.read_excel("train_dataset/yt/yt2.xlsx")
yt2_df
```

```python
yt_df = yt1_df.merge(yt2_df, how="inner", on="image")
yt_df
```

```python
yt_df.rename({"Просмотры за месяц": "views", "Количество подписчиков": "subscribers"}, axis=1, inplace=True)
yt_df
```

```python
yy1 = pd.DataFrame(columns=['text', 'y'])
yy2 = pd.DataFrame(columns=['text', 'y'])
```

```python
files = yt_df['image'].tolist()
path = "train_dataset/yt/yt1/"
for file in tqdm(files):
    if file.lower().endswith(ALLOWED_FILES):
        try:
            text = " ".join(reader.readtext(f"{path}/{file}", detail=0, paragraph=True))
            y1 = yt_df[yt_df['image'] == file]['subscribers'].values[0]
            y2 = yt_df[yt_df['image'] == file]['views'].values[0]
            d1 = {"text": text, "y": y1}
            d2 = {"text": text, "y": y2}
            yy1 = yy1._append(d1, ignore_index=True)
            yy2 = yy2._append(d2, ignore_index=True)
        except:
            continue
```

```python
yy1
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(yy1.drop(['y'], axis=1), yy1['y'],
                                                    test_size=0.4,
                                                    random_state=228)
x_train
```

```python
text_features = ['text']
```

```python
model = CatBoostRegressor(iterations=5000, loss_function='MAE', learning_rate=0.01, max_depth=5, eval_metric="MAE",
                          leaf_estimation_backtracking="AnyImprovement", subsample=0.3, bagging_temperature=0.4,
                          langevin=True)
```

```python
model.fit(x_train, y_train, use_best_model=True, eval_set=(x_test, y_test), text_features=text_features, early_stopping_rounds=500)
```

```python
model.save_model("yt1_model")
```
